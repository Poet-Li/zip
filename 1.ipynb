```python
# 计算cost（每个路段的分段数自己设定）
def compute_cost(l, k, p): 
    cost = 0
    for i in range(len(l)):
        # duan = i * 4 + 4
        # duan = (i+1)**2 + 4
        duan = (4-i)**2 + 4
        cost += (l[i]/duan)**2*duan*k[i]/p[i]
    return cost

#产生初始解
def gen_ordinary():
    a = np.random.dirichlet(np.ones(4),size=1)
    p = [i for j in a for i in j]
    return p

#产生新解
def gen_new_p(p, t):
    normal_y = []
    while(len(normal_y) < 4):
        a = random.normalvariate(0, 1)
        if(a > 0):
            normal_y.append(a)
    summ = 0
    for i in range(4):
        summ += normal_y[i]**2
    summ = np.sqrt(summ)
    for i in range(4):
        normal_y[i] /= summ
    summ = 0
    for i in range(4):
        p[i] = p[i] + t * normal_y[i]
        summ += p[i]
    for i in range(4):
        p[i] /= summ
    return p
```

```python
# 染色体算法
maxgen = 1000001
def find_the_nearestpoint(x1, y1, edges_, lo, li):
    minn = np.inf
    minpoint = edges_[0][0]
    for edge_ in edges_:
        for two in edge_:
            dist = getdist(x1, y1, lo[two-1], li[two-1])
            if(dist == 0):
                continue
            if(dist < minn):
                minn = dist
                minpoint = two
    return minpoint

    
# 计算他们的loss
total_loss = np.zeros(5)
paths = [[[89,89]]] * 5
for u in range(1, 7):
    print("开始处理优先级为{}的道路".format(7-u))
    # 取出所有当前优先级的路
    road = links.loc[links["Importance level"] == 7-u]
    indexes = []
    edge = []
    for index, row in road.iterrows():
        indexes.append(index)
        start = row["Start node"]
        end = row["End node"]
        edge.append([start, end])
    edge = np.array(edge)
    indexes = np.array(indexes)

    # 生成5条染色体，每条染色体的非0元素即为该染色体锁包含的边序号
    length = len(edge)
    shuffle = np.random.permutation(length)
    chrome = np.zeros((5, length))
    j = 0
    for i in range(length):
        chrome[j][i] = indexes[shuffle[i]]
        j += 1
        j %= 5


    path = [[0,0]] * 5
    loss = np.zeros(5)
    loss.fill(np.inf)
    recordplot = []
    for g in range(maxgen):
        if(g % 10000 == 1):
            s = time.time()
        #一条一条处理染色体
        # path = [[]] #5行，每行记录路径
        loss1 = np.zeros(5)
        for i in range(5):
            chrome_rowi = chrome[i]
            #print(chrome_rowi)
            nonzero_indexs = np.nonzero(chrome_rowi)[0] # 第i行中非0元素的索引
            edge_indexs = indexes[nonzero_indexs] # 第i行包含的边在总集合中的序号
            edge_curr = edge[nonzero_indexs] # 得到第i行包含的具体的边
            currpoint = paths[i][-1][-1]
            # print("currpoint:",currpoint)
            # loss1.fill(np.inf)
            # 循环一行，得到最短路径的loss

            path1 = []
            for _ in range(len(edge_curr)):
                minpoint = find_the_nearestpoint(lo[currpoint-1], li[currpoint-1], edge_curr, lo, li)
                #print("minpoint:", minpoint)
                path1.append([currpoint, minpoint])
                loss1[i] += f[currpoint-1][minpoint-1]
                b = np.where(edge_curr == minpoint)
                #print("b:", b)
                adjpoint = edge_curr[b[0][0]][1^b[1][0]]
                loss1[i] += f[minpoint-1][adjpoint-1]
                #print("adjpoint:", adjpoint)
                path1.append([minpoint, adjpoint])
                currpoint = adjpoint
                edge_curr = np.delete(edge_curr, b[0][0], 0)
            #如果比原来更优，则更新
            #print("i:{},loss1:{}".format(i,loss1))
            if(loss1[i] < loss[i]):
                loss[i] = loss1[i]
                path[i] = path1.copy()
                #if(loss1[i] == 0):
                    #print("path[i]:", path[i])
        
        #print("loss:", loss)
        maxindex = np.argmax(loss)
        minindex = np.argmin(loss)
        #print("maxindex",maxindex)
        #print("minindex", minindex)
        valid_index = np.nonzero(chrome[maxindex])
        r = random.randint(0, len(valid_index[0])-1)
        r = valid_index[0][r]
        #print("r:",r)
        tmp = chrome[maxindex][r]
        chrome[maxindex][r] = chrome[minindex][r]
        chrome[minindex][r] = tmp
        #print(chrome[maxindex][r] - tmp)
        
        #print(loss)

        chrome_rowi = chrome[minindex]
        #print(chrome_rowi)
        nonzero_indexs = np.nonzero(chrome_rowi)[0] # 第i行中非0元素的索引
        edge_indexs = indexes[nonzero_indexs] # 第i行包含的边在总集合中的序号
        edge_curr = edge[nonzero_indexs] # 得到第i行包含的具体的边
        currpoint = paths[i][-1][-1]
        # print("currpoint:",currpoint)
        # loss1.fill(np.inf)
        # 循环一行，得到最短路径的loss
        loss12 = 0
        path1 = []
        for _ in range(len(edge_curr)):
            minpoint = find_the_nearestpoint(lo[currpoint-1], li[currpoint-1], edge_curr, lo, li)
            #print("minpoint:", minpoint)
            path1.append([currpoint, minpoint])
            loss12 += f[currpoint-1][minpoint-1]
            b = np.where(edge_curr == minpoint)
            #print("b:", b)
            adjpoint = edge_curr[b[0][0]][1^b[1][0]]
            loss12 += f[minpoint-1][adjpoint-1]
            #print("adjpoint:", adjpoint)
            path1.append([minpoint, adjpoint])
            currpoint = adjpoint
            edge_curr = np.delete(edge_curr, b[0][0], 0)

        loss[minindex] = loss12
        path[minindex] = path1.copy()
        #print(loss)
        #print('---')
        if(g % 10000 == 0):
            e = time.time()
            print("第{}次循环完毕，用时{:.3f}s，loss为{}".format(g, e-s, np.max(loss)))
            recordplot.append([g, np.max(loss)])
    total_loss += loss    
    for i in range(5):
        paths[i] += path[i]
```
